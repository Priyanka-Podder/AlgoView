```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

# Open the image
image = Image.open('example.jpg')  # Replace with your image path
# Ensure the image is in RGB mode
image = image.convert("RGB")
# Initialize ImageDraw
draw = ImageDraw.Draw(image)
# Define the text and font
text = "Your Text Here"
font = ImageFont.truetype("font.ttf", size=50)  # Replace with your font path and size
# Position of the text
position = (50, 50)  # Adjust as needed
# Add text to the image (using black color)
draw.text(position, text, font=font, fill="black")
# Display the result image
plt.imshow(image)
plt.axis('off')  # Hide axis
plt.show()
```
